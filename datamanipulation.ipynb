```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/krishnaik06/Complete-Python-Bootcamp/refs/heads/main/10-Data%20Analysis%20With%20Python/data.csv")
df.head()
```

```python
df.dtypes
```

```python
df.describe()
```

```python
#handling missing values

df.isnull()
```

```python
df.isnull().sum()
```

```python
df.isnull().any()
```

```python
df_filled = df.fillna(0)
```

```python
#filling missing values with mean of the column

df['Sales_fillNA'] = df['Sales'].fillna(df['Sales'].mean())
```

```python
df.head()
```

```python
#renaming columns

df = df.rename(columns={'Date':'Sale Date'})
df.head()
```

```python
#change data type

df['Value_new'] = df['Value'].fillna(df['Value'].mean()).astype(int)
df.head()
```

```python
df.dtypes
```

```python
df['New Value'] = df['Value'].apply(lambda x: x * 2)
df.head()
```

```python
#data aggregating and grouping

grouped_mean = df.groupby('Product')['Value'].mean()
grouped_mean
```

```python
grouped_mean = df.groupby(['Product', 'Region'])['Value'].sum()
grouped_mean
```

```python
grouped_mean = df.groupby(['Product', 'Region'])['Value'].mean()
grouped_mean
```

```python
#aggregate multiple functions

grouped_agg = df.groupby('Region')['Value'].agg(['mean', 'sum', 'count'])
grouped_agg
```

```python
#merging and joining dataframe

df1 = pd.DataFrame({'Key': ['A', 'B', 'C'], 'Value1':[1,2,3]})
df2 = pd.DataFrame({'Key': ['A', 'B', 'D'], 'Value1':[4,5,6]})
```

```python
df1
```

```python
df2
```

```python
#merge dataframe on the key columns

pd.merge(df1, df2, on = 'Key', how = "inner")
```

```python

pd.merge(df1, df2, on = 'Key', how = "outer")
```

```python
pd.merge(df1, df2, on = 'Key', how = "left")
```

```python
pd.merge(df1, df2, on = 'Key', how = "right")
```
